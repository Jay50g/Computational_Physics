```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
# characteristics of the system
k=0.1
m=0.2
owega=np.sqrt(k/m)
```

```python
# X is the state of the system. X[0]= displacement, X[1]=velocity
```

```python
def f(t,X):   # system dynamics
    return [X[1], -k/m*X[0]]
```

```python
ans=sp.integrate.solve_ivp(f,(0,100),[0.2,0],dense_output=True)
```

```python
t=np.linspace(0,20,500)
```

```python
solution=ans.sol(t)
```

```python
solution.shape
```

```python
plt.plot(t,solution[0],'r-',lw=3)
plt.plot(t,solution[1],'g-',lw=3)
plt.grid()
plt.legend(['displacement','velocity'])
```
