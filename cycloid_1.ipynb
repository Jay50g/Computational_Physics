```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
omega=5
t=np.linspace(0,100,3000)
r=0.5
```

```python
x=r*omega*t+r*np.cos(omega*t)
y=-r*np.sin(omega*t)
plt.plot(x,y,'r-',lw=3)
plt.xlim([0.5,4])
```
