```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
g=9.8
l=1
ang=60
x0=np.array([np.radians(ang),0])
```

```python
def f(t,x):
    #return np.array([x[1],-g/l*np.sin(x[0])])
    theta,omega=x
    return np.array([omega,-g/l*np.sin(theta)])

def f2(t,x):
    return np.array([x[1],-g/l*x[0]])
```

```python
ans1=sp.integrate.solve_ivp(f,[0,100],x0,dense_output=True)
ans2=sp.integrate.solve_ivp(f2,[0,100],x0,dense_output=True)
```

```python
t=np.linspace(0,50,500)
X1=ans1.sol(t).T
X2=ans2.sol(t).T
```

```python
plt.plot(t,X1[:,0],'g-',lw=3)
plt.plot(t,X2[:,0],'r-',lw=1)
```
