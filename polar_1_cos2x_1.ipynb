```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
theta=np.linspace(0,2*np.pi,5000)
r=2*np.cos(2*theta)
```

```python
x=r*np.cos(theta)
y=r*np.sin(theta)
```

```python
c=['r-','g-','b-','m-']
```

```python
for i in range(4):
    plt.plot(x[i*len(x)//4:(i+1)*len(x)//4],
             y[i*len(x)//4:(i+1)*len(x)//4],
             c[i],lw=5)
    plt.axis('equal')
    plt.grid(alpha=0.4)
```
