```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=500
r=np.linspace(0,10,n)
p=np.linspace(0,5*np.pi,n)
```

```python
R,P=np.meshgrid(r,p)
```

```python
x,y=R*np.cos(P),R*np.sin(P)
z=P
```

```python
fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(projection='3d')
ax.plot_surface(x,y,z,cmap='brg',edgecolor='g')
```

```python
t=np.linspace(0,2*np.pi,n)
y=np.cos(t)
y2=y**2
plt.plot(t,y,'r-',lw=3,label='mimic AC electricity')
plt.plot(t,y2,'b-',lw=3,label='mimic AC energy')
plt.legend()
plt.grid(alpha=0.4)
```

```python
particles=np.linspace(0,np.pi/2,50)
```

```python
prediction=np.sqrt((np.cos(particles)**2).mean())
prediction
```
