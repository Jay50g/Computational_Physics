```python
import numpy  as np
import scipy as sp
import matplotlib.pyplot  as plt
```

```python
t=np.linspace(0,2*np.pi,3000)
```

```python
x=np.cos(t)
y=np.cos(t)
```

```python
plt.plot(x,y,'r-',lw=3)
```

```python
len(x)//4
```

```python
c=['r-','g-','b-','m-']
fig=plt.figure(figsize=(15,10))
for i in range(6):
    x=np.cos(t)
    y=np.sin(t+i*np.pi/6)
    ax=fig.add_subplot(2,3,i+1)
    ax.axis('equal')
    for j in range(4):
        ax.plot(x[j*len(x)//4:(j+1)*len(x)//4],y[j*len(x)//4:(j+1)*len(x)//4],c[j],lw=3)
```
