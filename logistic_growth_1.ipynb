```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
x0=100
xM=10e9
k=0.6
```

```python
def f(t,x):
    return np.array(k*x*(1-x/xM))
```

```python
sol=sp.integrate.solve_ivp(f,[0,100],[x0],dense_output=True)
```

```python
t=np.linspace(0,50,3000)
X=sol.sol(t)
```

```python
X.shape
```

```python
plt.plot(t,X[0],'r-',lw=3)
plt.grid(alpha=0.3)
```
